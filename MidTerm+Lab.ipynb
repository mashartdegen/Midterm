## Preparing & Modeling the Data

```python
# a stacked bar plot with errorbars

import numpy as np
import matplotlib.pyplot as plt

N = 10
viewMeans = (4.3, 7.1, 3.2, 2.9, 6.7, 2.3, 3.1, 4.9, 2.4, 14.8)
gdpMeans = (11.3, 5.4, 2.6, 6.3, 2.9, 2, 3.5, 9.1, 1.6, 7.3)
ind = np.arange(N) # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence
p1 = plt.bar(ind, viewMeans, width, color='#d62728')
p2 = plt.bar(ind, gdpMeans, width,
             bottom=viewMeans)
plt.ylabel('Percentage of WC Viewers')
plt.title('Scores by Viewership and GDP-Weighted Viewership')
plt.xticks(ind, ('US', 'Braz', 'Mex', 'Germ', 'Indo', 'Tur', 'Russ', 'Japan', 'Thai', 'Chi'))
plt.yticks(np.arange(0, 30, 2))
plt.legend((p1[0], p2[0]), ('Viewership', 'GDP-Weighted Viewership'))

plt.show()
```

## Analysis

- Brazil having high viewership percentage makes sense because they hosted the tournament
- The US and Japan had the highest GDP-weighted viewership percentages at the 2014 World Cup
- China, Indonesia, Turkey, and Thailand had very high viewership percentages in spite of the fact that their national teams did not qualify

## Preparing & Modeling the Data

```python
# these values will be along the x-axis
import matplotlib.pyplot as plt
# now I have imported the python function called matplotlib
# in this bracket, I will have my coordinates
plt.plot([], [], 'bo')
# the first list includes the Xs and the second has the Ys
X = [5, 11, 5, 18, 15, 2, 6, 2, 2, 4, 8, 10, 1, 3, 7, 6]
# now I will input the Y values of police shootings
Y = [4, 7, 4, 7, 7, 3, 4, 3, 3, 4, 7, 5, 3, 3, 4, 3]
plt.axis([0, 20, 0, 8])
# this is my graphs scale (it is slightly unusual but it fits the data well)
plt.ylabel('# of Games Played (Success)')
plt.xlabel('# Goals Scored')
# labels for the axes
plt.show()
# now I want to find the line of regression and plot it
def best_fit(X, Y):

    xbar = sum(X)/len(X)
    ybar = sum(Y)/len(Y)
    n = len(X) # or len(Y)

    numer = sum([xi*yi for xi,yi in zip(X, Y)]) - n * xbar * ybar
    denum = sum([xi**2 for xi in X]) - n * xbar**2

    b = numer / denum
    a = ybar - b * xbar

    print('best fit line:\ny = {:.2f} + {:.2f}x'.format(a, b))

    return a, b

# solution
a, b = best_fit(X, Y)

# plot points and fit line
plt.scatter(X, Y, c='brown')
yfit = [a + b * xi for xi in X]
plt.plot(X, yfit, c='red')
# now I need to find the correlation coefficient to see the strength of the line of best fit
import numpy as np
# importing the numpy program
np.corrcoef(X, Y)[0, 1]
# 0-1 is the range that r can lie in
# r can be positive or negative but the relationships I'm plotting are all positive
```
